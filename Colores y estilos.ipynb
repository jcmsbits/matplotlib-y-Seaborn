Matlab -> Se pasan en un s√≥lo argumento el color, los puntos y las lineas

```python
plt.style.use('dark_background')
fig, axes = plt.subplots(figsize = (8,8))
axes.plot(x, x+1,'r--')
axes.plot(x, x+2, 'b-')
axes.plot(x, x+3,'y.-')
axes.plot(x, x+4,'go:')
```

Pyplot -> Los colores (se pueden pasar como RGB), lineas y marcas se pasan como par√°metros.

```python
fig, axes = plt.subplots(figsize = (8,8))
# alpha le da transparencia
axes.plot(x, x+1, color = 'red', alpha = 0.2, linewidth = 8, linestyle = '-', marker = '8', markersize = 20, markerfacecolor = 'white')
axes.plot(x, x+2, color = 'blue', linewidth = 2, linestyle = '--', marker = 'D', markersize = 5, markerfacecolor = 'green')
axes.plot(x, x+3, color = '#66FF66', linewidth = 4, linestyle = 'dashed', marker = 'P', markersize = 20, markerfacecolor = 'gray')
axes.plot(x, x+4, color = '#CCFFCC', linewidth = 6, linestyle = ':', marker = 's', markersize = 11, markerfacecolor = 'blue')
```
