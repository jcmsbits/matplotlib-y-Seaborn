```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5, 11)
```

```python
y = np.sin(x)
```

Los label en los gráficos corresponde a la leyenda
Y los signos de peso hacen que salga con notación matemática

```python
fig, ((ax1, ax2)) = plt.subplots(1,2, figsize = (10,4))
ax1.plot(x, y, label = "$sin(x)$")
ax1.set_title("Relacion entre X ----- Y")
ax1.set_xlabel("X")
ax1.set_ylabel("Y")
ax1.legend()
ax2.plot(y, x)
ax2.set_title("Relacion entre Y ----- X")
ax2.set_xlabel("Y")
ax2.set_ylabel("X")
```

Con el argumento loc en legend específicamos la posición de la leyenda
y con bbox_to_anchor específicamos su posición en coordenadas

```python
plt.plot(x, y, color = 'r', marker = 'D', label = "sen(x)")
plt.title("Relación entre X ---- Y")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend(loc = "lower left", bbox_to_anchor = (0.7,0.5))
```
