```python
import matplotlib.pyplot as plt 
import seaborn as sns
import numpy as np
```

```python
x = np.linspace(0,5, 20)
x
```

```python
y = x ** 2
```

Plot es un gráfico de Línea

```python
plt.plot(x,y)
plt.show()
```

El tercer parámetro es el color, o se puede pasar directamente con el párametro "color = "r""
Colores:
y -> yellow
b -> blue
g -> green

Los makers es la forma que se cada punto
o -> O
x -> X
s -> []
D -> Diamanates

```python
plt.plot(x,y, "bX")
plt.show()
```

```python
plt.plot(x,y, "ro")
plt.show()
```

Union entre líneas:
Rayas -> -
Dos puntos -> :

```python
plt.plot(x,y, "ys-")
plt.show()
```

```python
plt.plot(x,y, "yD:")
plt.show()
```
