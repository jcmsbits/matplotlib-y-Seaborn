```python
import matplotlib.pyplot as plt 
import seaborn as sns
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
fig = plt.figure()
axes = fig.add_axes([0.5,0.1,0.5,0.9])
axes.plot(x,y, 'bx')
plt.show()
```

```python
fig = plt.figure()
axes_1 = fig.add_axes([0.1,0.1,0.5,0.9])
axes_2 = fig.add_axes([0.6,0.1,0.5,0.9])
axes_1.plot(x,y, 'bx')
axes_2.pie(y)
plt.show()
```

```python
fig = plt.figure()
axes_1 = fig.add_axes([0.1,0.1,0.8,0.9])
axes_2 = fig.add_axes([0.2,0.55,0.4,0.3])
axes_1.plot(x,y, 'bx')
axes_2.plot(y,x,'r')
axes_2.set_facecolor('gray')
plt.show()
```
