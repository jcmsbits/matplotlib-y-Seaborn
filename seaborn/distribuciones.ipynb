```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tip = sns.load_dataset("tips")
```

```python
tip.tail()
```

```python
sns.histplot(data=tip, x = 'tip', bins=15, cumulative= True)
plt.show()
```

Stat => 'density', 'probability', 'percent', 'count'

```python
sns.histplot(data=tip, x = 'tip', bins=15, cumulative= False, hue='sex', stat='percent')
plt.show()
```

multiple => 'stack', 'dodge', 'fill'

```python
sns.histplot(data=tip, x = 'tip', bins=15, cumulative= False, hue='sex', stat='count', multiple='stack')
plt.show()
```

```python
sns.histplot(data=tip, x = 'tip', bins=15, cumulative= False, hue='sex', stat='percent', multiple='dodge')
plt.show()
```

```python
sns.histplot(data=tip, x = 'tip', bins=15, cumulative= False, hue='sex', stat='count', multiple='fill')
plt.show()
```

```python
sns.kdeplot(data = tip, x = 'tip', hue='sex')
plt.show()
```

```python
sns.kdeplot(data = tip, x = 'tip', hue='sex', cumulative= False, fill= True, bw_adjust=1)
plt.show()
```

```python
sns.ecdfplot(data=tip, x = 'tip', hue='sex', stat = 'count')
plt.show()
```

```python
sns.displot(data=tip, x = 'tip', hue='sex', kind='kde', multiple = 'stack')
plt.show()
```

```python
sns.displot(data=tip, x = 'tip', hue='smoker', kind='hist', multiple = 'dodge')
plt.show()
```

```python
sns.displot(data=tip, x = 'tip', hue='sex', kind='ecdf', stat = 'count')
plt.show()
```
