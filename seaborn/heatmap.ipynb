```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
tips.tail(5)
```

```python
tips_corr = tips.corr(numeric_only=True)
```

```python
sns.heatmap(data = tips_corr, annot=True, cmap="coolwarm", linewidths=5, linecolor='black', vmin = -1, vmax=1)
```

```python
sns.heatmap(data = tips_corr, annot=True, cmap="coolwarm", linewidths=5, linecolor='black', vmin = 0, vmax=1, cbar=False)
```

```python
sns.heatmap(data = tips_corr, annot=True, cmap="coolwarm", linewidths=5, linecolor='black', vmin = 0.5, vmax=1, cbar=True)
```
