```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.tail(8)
```

```python
sns.jointplot(data=tips, x="total_bill", y = "tip", hue = "sex", kind='hist')
plt.show()
```

```python
plt.figure(figsize=(8,8))
sns.jointplot(data=tips, x="total_bill", y = "tip", hue = "sex", kind='kde')
plt.show()
```

```python
sns.jointplot(data=tips, x="total_bill", y = "tip", hue = "sex", marginal_ticks=True)
plt.show()
```

```python
kws = dict(bins=25, dodge = True, fill = True)
```

```python
sns.jointplot(data=tips, x="total_bill", y = "tip",
              hue = "sex",
              marginal_ticks=True, 
              marginal_kws={"fill":True,"multiple":"stack"})
plt.show()
```

```python
sns.pairplot(data=tips, hue='sex', palette='coolwarm', corner= True, kind= 'scatter')
plt.show()
```
