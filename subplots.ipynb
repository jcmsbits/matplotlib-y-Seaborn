```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

```python
y = np.sin(x)
```

```python
fig, axes = plt.subplots()
axes.plot(x,y, 'b')
plt.show()
```

Con subplots me devuelve un array en axes en el cual puedo hacer referencia a cada gráfico

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y,'b')
axes[1].plot(y,x, 'y', marker = 'o')
```

```python
fig, (axes_1, axes_2) = plt.subplots(nrows=1, ncols=2)
axes_1.plot(x,y, color = 'g', marker = 's')
axes_2.pie(x)
plt.show()
```

No hace falta específicar nrows y ncols en los argumentos. El primero son las filas y la segunda las columnas

```python
fig, axes = plt.subplots(2,4)

axes[0,0].plot(x, y, color = 'g')
axes[0,1].plot(y, x, color = 'y', marker = 's')
axes[0,2].plot(x, np.cos(x), 'b', marker = 'd')
axes[0,3].plot(x, np.tan(x), 'r', marker = 'o')
fig.tight_layout()
```

```python
fig, ((ax1, ax2,ax3,ax4), (ax5,ax6,ax7,ax8)) = plt.subplots(2,4)
ax1.plot(x,y,'rD:')
ax2.plot(y,x, 'gs-')
ax3.plot(y, np.tan(y), 'bs:')
ax4.plot(x, np.log2(x))
ax5.plot(y, np.cos(y), color = 'r', marker = 's')
ax6.plot(x, np.cos(x) ** 2, color = 'g', marker = 'D')
ax7.pie(x)
ax8.hist(x)
fig.tight_layout()
```
